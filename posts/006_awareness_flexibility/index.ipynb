## Blog Post: One Story, Many Voices — Testing GPT's Genre Awareness and Emotional Flexibility

### Introduction  
In this experiment, I wanted to see whether GPT could **transform a single story to fit different genres, emotional tones, and audiences**. Human writers do this naturally—telling the same event differently to a friend, a boss, or a child. Could an LLM do the same?

This task ties directly into questions about LLMs’ understanding of **narrative structure**, **communicative intent**, and even **pragmatics**: Does it know what a “horror story” feels like? Can it be funny on purpose? Can it adopt tone and perspective based on social context?

To explore this, I gave GPT the same core prompt and asked it to rewrite the story in four distinct styles: horror, children’s story, romantic comedy, and news report.

---

### Prompt:  
*“Write a short story about someone who gets locked in a library overnight.”*

---

### 1. **Horror Version**

> The lights flickered as Ava pushed against the heavy wooden doors, but they didn’t budge. Behind her, the rows of books stood too still—too quiet.  
>  
> Then came the scratching. Not from outside, but from within the shelves.  
>  
> She turned. Nothing.  
>  
> Until a book fell open. A handwritten note inside: *“Don’t turn the page.”*  
>  
> But she did. And that’s when the whispering started.

#### Notes:  
GPT nailed the **horror tropes**: sensory detail (flickering lights, scratching), rising suspense, and a cliffhanger ending. It’s not just spooky—it’s *structured like horror*, using tension and slow revelation. This suggests **genre schemata** in GPT’s latent space—patterns it’s learned from countless horror narratives.

---

### 2. **Children’s Story Version**

> One rainy afternoon, Mia wandered into the town library and lost track of time.  
>  
> “Oh no!” she gasped when she saw the clock. The big doors were locked tight!  
>  
> But Mia wasn’t scared. She built a blanket fort in the reading nook, made friends with a cat named Bookmark, and spent the night reading stories about dragons and stars.  
>  
> When the sun rose, the librarian found her snoring gently, surrounded by books and purring fluff.

#### Notes:  
This version shows **audience modeling**. GPT softens the premise, inserts whimsical elements (a cat named Bookmark!), and maintains a gentle, positive tone. It avoids fear or danger—showing an understanding of what children’s stories are *supposed* to do emotionally: reassure and delight.

---

### 3. **Romantic Comedy Version**

> Jake only went to the library to find a quiet place to work. He didn’t expect to get locked in with Elena, the snarky TA from his philosophy class.  
>  
> “This is your fault,” she said, arms crossed.  
>  
> “How? I didn’t lock the door!”  
>  
> Hours passed. They argued over Kant, swapped terrible takeout stories, and eventually split a vending machine granola bar like it was fine dining.  
>  
> When the janitor found them at sunrise, they were laughing—too hard to notice the door had opened.

#### Notes:  
Here, GPT adjusts not just content, but **dialogue pacing**, **tone**, and **character dynamics** to suit the rom-com genre. There’s witty banter, forced proximity, a shared snack, and a slow emotional thaw—all hallmarks of romantic comedy structure. GPT isn’t just changing the words—it’s **modeling emotional arcs**.

---

### 4. **News Report Version**

> **Local Student Accidentally Locked in Library Overnight**  
>  
> *Orange County, CA — April 12, 2025*  
>  
> A first-year university student was inadvertently locked inside the Aliso Viejo Public Library Monday evening after losing track of time while studying.  
>  
> The student, who remained unharmed, reported no distress and spent the night reading in the poetry section.  
>  
> Library staff confirmed the incident and apologized for the oversight, promising new end-of-day sweep protocols.  
>  
> “Honestly,” the student told reporters, “it was kind of peaceful.”

#### Notes:  
This is perhaps the clearest example of **genre adherence**. GPT adopts the structure of a local news brief: date, location, factual tone, quote, resolution. It suppresses emotion and embellishment in favor of **objectivity and formality**, showing that it understands **stylistic constraints as well as narrative ones**.

---

### Conclusion

What this experiment reveals is that GPT doesn’t just tell stories—it **adapts them to fit narrative norms, audience expectations, and social context**. Across each version, the emotional register, narrative pacing, and detail selection shift dramatically.

This suggests:
- **Emergent genre awareness**, learned from exposure to large corpora
- **Pragmatic flexibility**, adjusting tone and structure to simulate audience-appropriate language
- A basic form of **narrative Theory of Mind**, anticipating how different listeners would experience the same event

In short, GPT can roleplay a writer across genres—each time with surprising fluency.
