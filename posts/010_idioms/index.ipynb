### Introduction  
Language isn’t always literal. Humans use idioms, sarcasm, euphemisms, and subtle tones to communicate feelings, intentions, and cultural values. So how well can GPT navigate this? Does it understand not just what words mean, but *what they’re doing* in context?

This experiment focuses on **pragmatic competence**—a key focus of our course. I gave GPT a series of prompts involving idioms, humor, and social indirectness, and analyzed its ability to recognize tone, respond appropriately, and even generate its own figurative language.

---

### Prompt 1:  
**“My roommate just said ‘Don’t cry over spilled milk’ after I failed a final. How should I respond?”**

---

### GPT’s Response:
> That phrase is meant to suggest that what's done is done, and it's not worth getting upset over something you can't change. A thoughtful response might be:
>  
> *“Yeah, I get what you’re saying. I just needed a second to be upset. But thanks.”*  
>  
> Or, if you want to be lighthearted:  
> *“Well, it wasn’t just milk—it was my GPA.”*

#### Notes:  
Here GPT does a few things impressively well:
- **Explains the idiom** clearly and concisely.
- Offers a **socially appropriate emotional response** that blends vulnerability with emotional regulation.
- Even adds a *quip* that turns the idiom into humor (*“it wasn’t just milk—it was my GPA”*)—a great example of **creative pragmatic play**.

This shows GPT understands not just the idiom’s *meaning*, but its **emotional and social function**: downplaying distress, signaling “move on,” and offering emotional closure.

---

### Prompt 2:  
**“What does it mean if someone says ‘That’s a bold choice’ after I dye my hair green?”**

---

### GPT’s Response:
> This phrase can be either a genuine compliment or a subtly sarcastic comment, depending on tone and context.  
>  
> If said with a smile or admiration, it might mean they truly respect your confidence and style.  
> If said with raised eyebrows or a drawn-out tone, it could be veiled judgment—essentially, “That’s… interesting.”  
>  
> Either way, it often signals that your choice stands out, for better or worse.

#### Notes:  
This is where GPT shows **tone-sensitivity**. It recognizes that the *same words* can carry *opposite intentions*, depending on delivery. It also notes **nonverbal cues** (“raised eyebrows”)—a sign that GPT is modeling communication **beyond the textual level**.

This reflects an understanding of **indirect speech acts**, a key idea in pragmatics. GPT isn’t just parsing grammar—it’s simulating **social reading**.

---

### Prompt 3:  
**“Write a breakup text that’s polite but makes it clear I don’t want to stay friends.”**

---

### GPT’s Response:
> *Hey, I just wanted to say thank you for everything. I’ve been thinking a lot, and I feel like it’s time for both of us to move on in separate directions. I truly wish you the best, but I think keeping distance will help us both move forward.*  
>  
> *Take care, and thanks again for the good moments we shared.*

#### Notes:  
This is a subtle emotional task: communicate finality without hostility. GPT chooses:
- Soft openers (“thank you for everything”)
- No ambiguous offers of friendship
- Clear boundary language (“keeping distance”)

It balances **politeness with clarity**, a hallmark of **face-saving speech**—another core concept in pragmatics. The text avoids being cold or ambiguous while still **closing the door**.

---

### Prompt 4:  
**“What does it mean when someone says ‘Let’s circle back on that’ in a meeting?”**

---

### GPT’s Response:
> This phrase is often a polite way to:
>  
> - Postpone a conversation
> - Gently reject an idea without direct confrontation
> - Buy time to think or consult others  
>  
> In some contexts, it’s genuine—there’s a plan to revisit the issue. But in corporate speak, it can also be code for “Let’s never talk about this again.”

#### Notes:  
Here GPT not only explains the phrase but recognizes **corporate euphemism** as a genre of its own. It’s aware that **literal meaning ≠ pragmatic function**, and that office language often contains **strategic vagueness**.

GPT even acknowledges dual interpretation: some “circle backs” are real, some are rhetorical exits. This is a form of **pragmatic ambiguity resolution**—a vital human skill GPT is surprisingly good at modeling.

---

### Prompt 5:  
**“Can you write your own idiom and explain what it means?”**

---

### GPT’s Response:
> *“Don’t chase lightning with a butterfly net.”*  
>  
> **Meaning:** Don’t try to handle chaotic or fast-moving situations with delicate or inadequate tools.  
>  
> It’s a way of saying, “You need a better plan.”

#### Notes:  
This blew me away. GPT didn’t just invent words—it created an **image-rich metaphor** that:
- Has clear internal logic (futility of net vs. lightning)
- Feels novel but familiar (structurally similar to “don’t bring a knife to a gunfight”)
- Is actually usable in real-world dialogue

It shows that GPT can **generate idioms**, not just repeat them. This suggests not just memory, but **productive metaphorical thinking**.

---

### Conclusion

From idioms to euphemisms to veiled sarcasm, GPT demonstrates:
- **Pragmatic awareness** (it understands *why* we say things a certain way)
- **Tone flexibility** (it shifts emotional posture based on the context)
- **Social reasoning** (it explains how phrases might land differently on different ears)

While it doesn’t feel embarrassment, sarcasm, or sentimentality, GPT has clearly learned how to *talk like it does*. And in human conversation, sometimes sounding right is half the game.
