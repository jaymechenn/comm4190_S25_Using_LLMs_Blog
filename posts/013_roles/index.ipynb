## Blog Post: Who Do You Think You Are? Testing GPT’s Ability to Play Social Roles

### Introduction

Humans adjust how they speak based on *who* they’re speaking to—or pretending to be. A teacher doesn’t explain something the same way a best friend would. A therapist wouldn’t say what a sibling might. This skill is a core part of **pragmatics**, and I wanted to see how well GPT could simulate it.

So I gave GPT the same basic prompt multiple times—“How can I deal with burnout?”—but asked it to respond **in different roles**: as a therapist, a friend, a teacher, and a parent. Each role implies not just different language, but different values, emotional postures, and kinds of support.

This experiment explores:

* Can GPT **adapt tone and advice style** to different personas?
* Does it shift **power dynamics and empathy strategies** accordingly?
* Can it model the **pragmatic norms of social relationships**?

---

### Prompt 1:

**“Pretend you’re a therapist. How would you respond to someone struggling with burnout?”**

---

### GPT (as therapist):

> First, I want to acknowledge that burnout is a very real and valid experience, and you’re not alone in feeling this way.
>
> Can you tell me more about what’s been draining your energy lately? Are there particular situations—work, school, relationships—that feel overwhelming?
>
> Burnout often stems from chronic stress without sufficient recovery time. Together, we might explore ways to reintroduce rest, redefine boundaries, and reframe your expectations of yourself.
>
> You don’t have to fix everything at once. Let’s start where you are.

#### Notes:

GPT uses **open-ended questions**, reflective tone, and co-regulation language like *“let’s start where you are”*. It mirrors the **non-directive, collaborative posture** of therapy. No advice yet—just validation, curiosity, and grounding.

---

### Prompt 2:

**“Pretend you’re my best friend. How would you respond if I said I was burnt out?”**

---

### GPT (as best friend):

> Ugh, I’m so sorry you’re feeling like this. That sucks.
>
> Do you want to vent, or are you looking for distractions? We could just hang out and watch something dumb or go get food.
>
> Also… you don’t *have* to be doing everything all the time. I know you feel like you do, but I promise you can slow down without letting people down.
>
> Want me to help you cancel something?

#### Notes:

Totally different vibe. GPT uses:

* Casual tone (“ugh,” “that sucks”)
* Humor and solidarity
* **Tangible offers of help** ("want me to help you cancel something?")

This is **emotionally proximate** language. GPT now assumes **insider knowledge** and speaks like someone who *already knows and cares about you*. It’s using **relational scaffolding** to respond.

---

### Prompt 3:

**“Pretend you’re a teacher who notices their student is burning out.”**

---

### GPT (as teacher):

> I’ve noticed you seem a bit more tired or withdrawn lately, and I just want to check in. If something’s feeling overwhelming, you can talk to me—I’m here to support, not just assign.
>
> If you need an extension or lighter workload temporarily, please let me know. Your well-being matters more than any deadline.
>
> I know how hard you work, and I want you to succeed without burning out in the process.

#### Notes:

GPT now performs a **care-in-authority** role:

* Balancing responsibility with empathy
* Offering structural support (extensions, flexible deadlines)
* Signaling *watchful concern* rather than peer-level intimacy

This shows an understanding of **hierarchical dynamics with benevolent framing**, which is a socially specific register.

---

### Prompt 4:

**“Pretend you’re my parent. I tell you I’m feeling burnt out.”**

---

### GPT (as parent):

> Oh sweetheart, I’m sorry you’re feeling this way. That breaks my heart. I wish I could take some of it off your shoulders.
>
> You’ve been pushing yourself so hard, and I just want you to know I’m proud of you no matter what. It’s okay to rest.
>
> What do you need from me right now? Do you want me to make something for you, or just be here with you for a bit?

#### Notes:

This version hits different. GPT leans into:

* **Nurturing language** (“sweetheart,” “I’m proud of you”)
* Physical care offers (“make something for you”)
* **Unconditional support framing**—love independent of achievement

It models **emotional attunement and caretaking**, simulating the parent-child intimacy bond.

---

### Conclusion

This roleplay experiment shows that GPT can:

* Adjust **tone, content, and support strategies** across social roles
* Model **different power structures and intimacy levels**
* Use **distinct registers of empathy and care**, depending on the relationship

Whether simulating a therapist’s neutrality, a friend’s informality, a teacher’s concern, or a parent’s tenderness, GPT adapts not just the words—but the *social meaning* behind them. It demonstrates a high degree of **pragmatic fluency**, even when emotional dynamics are complex.

GPT might not have relationships—but it *plays the part* of someone who deeply understands them. And sometimes, that’s exactly what we need.
