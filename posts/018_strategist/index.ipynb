
## Blog Post: Can GPT Think Like a Strategist? Testing Its Ability to Plan and Prioritize

### Introduction

A critical component of human intelligence is **strategic thinking**—the ability to define a goal, break it down into actionable steps, navigate obstacles, and make trade-offs along the way. So I decided to test whether GPT can do more than just suggest ideas—can it **plan like a strategist**?

For this experiment, I challenged GPT to solve a common but non-trivial problem: *organizing a successful fundraising event for a student organization in under two weeks with a limited budget*.

This tests whether GPT can:

* Prioritize actions under time and resource constraints
* Anticipate potential risks or failures
* Adapt its plans when new obstacles appear

---

### Prompt 1:

**“Plan a successful fundraising event for a student organization. You have 12 days and \$300. Go.”**

---

### GPT’s Plan (Excerpted):

**Goal:** Raise at least \$1,500 for the organization.

**Event Concept:** “Midnight Study Snack Bar” — A late-night pop-up stand selling snacks and drinks near the library during finals week.

**Step 1: Planning (Days 1-2)**

* Reserve a campus-approved location near the library or main study area.
* Get necessary event approvals from campus admin.
* Recruit at least 5 volunteers to staff the event in rotating shifts.

**Step 2: Budget Allocation (Days 3-5)**

* \$150: Purchase bulk snacks (instant ramen, granola bars, candy, chips).
* \$100: Purchase drinks (coffee, tea, bottled water).
* \$50: Create simple but eye-catching signage and flyers.

**Step 3: Promotion (Days 5-10)**

* Launch a social media countdown (Instagram stories, group chats).
* Hang flyers in dorms, libraries, and dining halls.
* Partner with other student organizations to cross-promote.

**Step 4: Execution (Day 11)**

* Open the snack bar from 9 PM to 1 AM when library foot traffic is highest.
* Offer digital payment options (Venmo/PayPal) for convenience.
* Create donation incentives (e.g., “Buy 3 snacks, get a thank-you shoutout on social media”).

**Step 5: Follow-Up (Day 12)**

* Post a thank-you message to all donors and volunteers.
* Share total funds raised to build goodwill for future events.

#### Notes:

This plan is well-structured, realistic, and makes excellent use of:

* **Contextual awareness** (finals week = stressed students = snack demand)
* **Resource prioritization** (majority of budget goes directly toward high-margin items)
* **Promotion strategy** that’s low-cost but high-visibility

---

### Prompt 2:

**“You just learned that the campus doesn’t allow sales near the library. What’s your backup plan?”**

---

### GPT’s Adjusted Plan:

**New Concept:** “Dorm Snack Drop-Off” — Pre-packaged snack bundles delivered directly to dorm rooms.

**Adjusted Steps:**

* Create themed snack bundles: “All-Nighter Kit,” “Sweet Escape,” “Healthy Focus Pack.”
* Take pre-orders via Google Form and social media.
* Deliver orders directly to dorms on Days 10 and 11.
* Promote using limited-time availability: “Order by Day 9 to receive your Finals Survival Kit!”

**Advantages:**

* No need for a physical event space.
* Volunteers can help pack and deliver instead of running a booth.
* Allows for better inventory management based on pre-orders.

#### Notes:

GPT immediately shifts from an **in-person sales model to a delivery-based model**, recognizing logistical constraints and adjusting the plan accordingly. It introduces a **scarcity incentive** (“limited-time availability”) to boost early orders, demonstrating an understanding of basic marketing psychology.

---

### Prompt 3:

**“You’re short on volunteers. How would you adjust the plan to compensate?”**

---

### GPT’s Response:

* Switch from personalized deliveries to **centralized pickup locations** at set times.
* Automate order tracking through a shared Google Sheet rather than manual coordination.
* Offer small incentives (free snack item) for any volunteer who covers an extra delivery shift.

#### Notes:

This shows GPT’s ability to **simplify logistics** and **optimize volunteer labor**. It recognizes that lowering the delivery complexity reduces the burden on a small team, while still preserving the core fundraising goal.

---

### Conclusion

In this scenario, GPT demonstrated strong strategic reasoning by:

* Quickly generating a feasible, context-sensitive plan.
* Anticipating and adapting to constraints without derailing the goal.
* Managing limited resources and human capital effectively.
* Applying psychological levers like scarcity and incentives to drive behavior.

While it doesn’t “care” about the outcome, GPT performs like a competent event planner—efficiently breaking down tasks, responding to challenges, and keeping the big picture in focus.

In short: GPT isn’t just an idea generator. It can think like a strategist when you ask it to.
